```python
#Tuning the Decision Tree: 
import numpy as np
from sklearn.metrics import mean_squared_error

from sklearn.model_selection import cross_val_score, KFold

target_col= "ARR Year Growth"

X= dummies.drop([target_col, "growth_percent", "Model "], axis = "columns")
y=dummies.loc[:,target_col]

kf = KFold(n_splits=len(X), shuffle=True)

dtr = DecisionTreeRegressor(max_depth=1, random_state=1)
scores = cross_val_score(dtr, X, y, cv=kf, scoring='neg_mean_squared_error')
mse = -np.array(scores).mean()
rmse = np.sqrt(mse)
rmse
```

```python
max_depth_range = range(1, 8)

rmse_scores = []

for depth in max_depth_range:
    dtr = DecisionTreeRegressor(max_depth=depth, random_state=1)
    neg_mse_scores = cross_val_score(dtr, X, y, cv=kf, scoring='neg_mean_squared_error')
    mse_scores = -np.array(neg_mse_scores)
    mse_ave_score = mse_scores.mean()
    rmse_ave_score = np.sqrt(mse_ave_score)
    rmse_scores.append(rmse_ave_score)
rmse_scores
```

```python
# Plot max_depth (x-axis) versus RMSE (y-axis).
fig, ax = plt.subplots()
ax.plot(max_depth_range, rmse_scores)
ax.set_xlabel('max_depth')
ax.set_ylabel('RMSE (lower is better)');
```
